```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Train.csv")
```

```python
df.head()
```

```python
df['date_time'] = df['date_time'].astype('datetime64[ns]') #Changing dtype from object to datetime64[ns]
```

```python
df['Year'] = pd.DatetimeIndex(df['date_time']).year #Adding Column of Year
df['Month'] = pd.DatetimeIndex(df['date_time']).month #Adding Column of Month
```

```python
df['Day'] = pd.DatetimeIndex(df['date_time']).day #Adding Column of day
```

```python
df['Hour'] = pd.DatetimeIndex(df['date_time']).hour #Adding Column of Hour
```

```python
df['is_holiday'][~df['is_holiday'].isin(['None'])] = 1 #Non - Holidays are marked as 1
df['is_holiday'][df['is_holiday'].isin(['None'])] = 0 #Holidays are marked as 0
```

```python
lst = []
for x in ['weather_type','weather_description']:
    unq = df[x].unique()
    k = 0
    dcu = {}
    for y in unq:
        dcu[y] = k # Mapping Weather Type & Weather Description
        k+=1
    lst.append(dcu)
    df[x] = df[x].map(dcu).astype(int)
```

```python
df.to_csv('New_train.csv', index=False) #Saving it to new CSV file
```
